```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

#Load job market
df = pd.read_csv('job-market.csv')
df.head(3)
```

```python
jobGroupByClassification = (
  df.groupby('Classification')['Classification']
    .count()
    .reset_index(name='Number of Postings')
    .sort_values(by=['Number of Postings'], ascending=False)
)
```

```python
sns.set(rc={"figure.figsize":(18, 12)})
(
  sns
    .barplot(
      data=jobGroupByClassification,
      x='Number of Postings',
      y='Classification',
    )
    .set(title='Jobs and Number of Postings')
)
```

```python
jobPostsBySalaryRange = df.assign(AverageSalary = (df['HighestSalary'] + df['LowestSalary']) / 2)
jobPostsBySalaryRange = jobPostsBySalaryRange.groupby(
  pd.cut(
    jobPostsBySalaryRange['AverageSalary'],
    bins=[0, 30, 40, 50, np.inf],
    labels=['0', '30', '40', '50']
  )
).size().reset_index(name='count')

jobPostsBySalaryRange.head(5)
```

```python
sns.set(rc={"figure.figsize":(6, 6)})
plt.pie(
    jobPostsBySalaryRange['count'],
    labels=jobPostsBySalaryRange['AverageSalary'],
    colors=sns.color_palette('pastel'),
    autopct='%.2f%%',
    # pctdistance=0.7
    
)
plt.title('Job Posts by Salary Range')

centreCircle = plt.Circle((0, 0), 0.5, fc='white')

fig = plt.gcf()
fig.gca().add_artist(centreCircle)
```
